Complex examples for column transformer

```python
# pipeline 1 for numerical features
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
# pipeline 1 for numerical columns
numerical_processor = Pipeline(steps = [('numerical_imputer',SimpleImputer(missing_values=np.nan,strategy = "mean")),
                                        ('standatardscaler',StandardScaler())])
numerical_processor
```

```python
#pipeline 2 for categorical columns
categorical_processor = Pipeline(steps = [('categorical_imputer',SimpleImputer(fill_value="missing", strategy = "constant")),
                                         ('onehot',OneHotEncoder(handle_unknown = "ignore"))])
categorical_processor
```

```python
# combining both pipeline using columntransformer
from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer([('Categorical',categorical_processor,["gender","city"]),('Numerical',numerical_processor,["age","Height","weight"])])
preprocessor
```

```python
# add an estimator(algorithm)
from sklearn.pipeline import make_pipeline
#make_pipeline(need to customize the pipleline)
final_pipe = make_pipeline(preprocessor, LogisticRegression())
final_pipe
```

```python
# return the prime numbers between or in the given range
def fun(num1,num2):
    li = []
    a = max(num1,num2)
    for i in range(num1,num2):
        if i > 1:
            for j in range(2,a):
                if i%j==0:
                    break
                else:
                    li.append(i)
    return li
num1,num2 = map(int,input().split())
print(fun(num1,num2))
```
